# TASK 5.2 Compare generated and original data

## Goal:  The double-date loss
- generated control cells/patients should be more similar to original control cells/patients than original diseased cells/patients **should be high**
- generated diseased cells/patients should be more similar to original diseased cells/patients than original control data **should be high**
- The difference between generated control and diseased cells/patients should be as close as possible to the difference between original control and diseased cells/patients. **should be low**

## Process new
1. Train a generator on each group seperatly on the original data.
2. Embedd the orignal data of each group  -> take each patients seperatly 
3. Generate patients -> embedd each patient using the same embedding space.
4. Get the KL-divergence between each pair of patient.
    1. the kl-divergence is measured on the meshgrid spanning the union of each pair of patient cells.
    - Problem: takes to long to do it for the pair. 
    2. Get two kernel density estimators, each trained on their respectiv patients embedding. Then use KL-divergence to compare the the estimations over the meshgrid.
5. Use the double-date metric too compare generators.



## Critisim
1. Factors that might affect the metric
    1. The embedding algorthim. The properties of some algorithm used will transfer to the metric. **Embedding groups in different space, will make the difference already big**
    2. The way KL-divergence is used. The span of the meshgrid will vary.
    3. The kernel density estimator. Different parameters might give different results.
    4. The generators themselves will vary with hyperparameters. (These are the models we want to validate.)
    5. The number of generated patients might affect metric. 
    6. The double-date metric might not be a good metric itself.

## Model selection techniques
Probabilistic Measures: Choose a model via in-sample error and complexity.
- Akaike Information Criterion (AIC).
- Bayesian Information Criterion (BIC).
- Minimum Description Length (MDL).
- Structural Risk Minimization (SRM).

appropriate when using simpler linear models like linear regression or logistic regression where the calculating of model complexity penalty (e.g. in sample bias) is known and tractable.

Resampling Methods: Choose a model via estimated out-of-sample error.
- Random train/test splits.
- Cross-Validation (k-fold, LOOCV, etc.).
- Bootstrap.

```python
def embedd_patients_with_tsne(df, embedder=None):
    """
    PARAMETERS
    ----------
    df : DataFrame 
    embedder : TSNEEmbedding 
    
    RETURN
    -------
    embedding : TSNEEmbedding 
    
    """
    if embedder is not None:
        return embedder.transform(df.values)
    else:
        tsne = TSNE(n_components=2, 
                perplexity=30, 
                metric="euclidean",
                random_state=42)

        tsneEmbedding = tsne.fit(df.values)
        return tsneEmbedding

def embedd_patients_with_umap(df, embedder=None):
    """
    PARAMETERS
    ----------
    df : DataFrame 
    embedder : umap embedding with existing embedded space
    
    RETURN
    -------
    embedding : df 
    
    embedder : UMAP embedder 
    
    """
    
    if embedder is not None:
        return embedder.transform(df)
    else:
        up = umap.UMAP(n_components=2, n_neighbors=10, random_state=0).fit(df)
        return up
```

### Find kl-divergence between all pairs of patients

```python
import CustomUtils as cu

class KL_LOSS():
    
    def __init__(self, dim_red=False):
        self.dim_red = dim_red
        
    def KL(self, a,b):
        """
        PARAMETERS
        ----------
        a : numpy array
        b : numpy array
        
        RETURNS
        -------
        kl divergence : int
            the nonsymmetric kulback liebler divergence between a and b. 
        """
        a = np.asarray(a, dtype=float)
        b = np.asarray(b, dtype=float)
        a = np.exp(np.where(a!= float('-inf'), a, 0))
        b = np.exp(np.where(b!= float('-inf'), b, 0))

        cond = np.logical_and(b != float('-inf'), a!= float('-inf'), b != 0)
    
        return np.sum(np.where(cond, a * np.log(a / b), 0))

    def KL_between_two_patients(self, KD1, KD2, symmetric=True):
        """
        PARAMETERS
        ----------
        p1 : numpy array
            the cells of the first patient
        p2 : numpy array
            the cells of the second patient
        symmetric : bool
            true if the kl divergence should be true.

        RETURNS
        -------
        kl divergence : float
            The symmetrci kulback liebler divergence measured on meshgrid spanning both p1 and p2. 
            The probability of each point is estimated with a KernelDensityEstimator trained on p1 and p2 separatly.
            The kl divergence is measured using these estimators.

        """

        if(symmetric):
            kl_score = self.KL(KD1, KD2) + self.KL(KD2, KD1)
        else:
            kl_score = self.KL(KD1, KD2)

        return kl_score 

    def get_cells(self, df, patient_id):
        """
        PARAMETERS:
        ----------
        df: DataFrame
            contains cell data.
        patient_id: int
            the patients id

        RETURNS
        -------
        patientCells: np.array 
            cells of patient patient_id
        """
        d = df.groupby("id").get_group(patient_id)
        return d[d.columns.difference(["id","group"])].values
       

    def crossexam(self, left_group, right_group, KD, name):
        """
        PARAMETERS
        ----------
        left_group : numpy array of unique values
        right_group : numpy array of unique values
        KD : density estimation for each individual: generated and control.
        name : the name of the cross examination, used for progressbar.

        RETURNS
        -------
        cross : DataFrame
            KL divergence scores between each individuals
        """

        left = np.sort(original[left_group].id.unique())
        right = np.sort(generated[right_group].id.unique())

        cross = pd.DataFrame(columns=left.astype("str"), index=right.astype("str"), dtype="int64")
        p = cu.ProgressBar(endVal=len(right)*len(left), name = name)
        while(p.move()):
            for i in right:
                for j in left:
                    cross.loc[str(i),str(j)] = self.KL_between_two_patients(KD["original"][left_group][j],\
                                                                      KD["generated"][right_group][i])
        return cross 

    def crossexam_same_data(self, group, group_name, KD, name):
        """
        PARAMETERS
        ----------
        group : numpy array of unique values
        group_name : str
        KD : density estimation for each individual: generated and control.
        name : the name of the cross examination, used for progressbar.

        RETURNS
        -------
        cross : DataFrame
            KL divergence scores between each individuals
        """

        c = "control"
        d = "diseased"

        left = np.sort(group["control"].id.unique())
        right = np.sort(group["diseased"].id.unique())


        cross = pd.DataFrame(columns=left.astype("str"), index=right.astype("str"), dtype="int64")

        p = cu.ProgressBar(endVal=len(right)*len(left), name = name)
        while(p.move()):
            for i in right:
                for j in left:
                    cross.loc[str(i),str(j)] = self.KL_between_two_patients(KD[group_name][c][j],\
                                                                      KD[group_name][d][i])
        return cross 

    def computeDensity(self, patients, grid):
        """
        PATIENTS
        --------
        patients: DataFrame
            contains cell data.

        RETURNS
        -------
        d: dictonary
            where the keys are the patients_ids and values are fitted KernelDensityEstimators.
        """
        patient_ids = np.sort(patients.id.unique())
        d = dict()
        estimation =dict()
        for p_id in patient_ids:
            #FFTKDE - Even though this is faster, it requires that the grid being evaluated contains all datapoints!
            d = TreeKDE(kernel = "epa", bw=1).fit(get_cells(patients, p_id))
            estimation[p_id] = d.evaluate(grid)

        return estimation
    
    def computeDensity_dr(self, df, grid):
        """
        PATIENTS
        --------
        patients: DataFrame
            contains cell data.
            
        RETURNS
        -------
        d: dictonary
            where the keys are the patients_ids and values are fitted KernelDensityEstimators.
        """
        patient_ids = np.sort(df.id.unique())
        d = dict()
        estimation = dict()
        for p_id in patient_ids:
            #FFTKDE - Even though this is faster, it requires that the grid being evaluated contains all datapoints!
            d = TreeKDE(kernel = "epa", bw=1).fit(self.get_cells(df, p_id))
            estimation[p_id] = d.evaluate(grid)
        
        return estimation

    def computeGrid(self, con, dis):
        """
        PARAMETERS
        ----------
        min_value : int
        max_value : int

        RETURNS
        -------
        mesh grid : numpy array
            a mesh grid between min_value and max_value
        """
        # Create a meshgrid that spans the range of the embedded orginal data
        max_value = max (con.max().max(), dis.max().max())
        min_value = min (con.min().min(), dis.min().min())
        grid_margins = [np.linspace(min_value-2, max_value+2, 120)] * 2
        grid = np.sort(np.stack(np.meshgrid(*grid_margins), -1).reshape(-1, len(grid_margins)), axis=0)
        return grid

    def computeGrid_dr(self, df, nr_points):
        """
        PARAMETERS
        ----------
        original : DataFrame, without id and group
        nr_points : int
        
        """
        # compute the max in each dimension
        max_val = df.max()
        min_val = df.min()
        #grid_margins = np.array([np.linspace(min_val-2, max_val+2, nr_points)] * len(df.columns))
        #grid = np.sort(np.stack(np.meshgrid(grid_margins, sparse=False), -1).reshape(-1, len(grid_margins)), axis=0)

        #print(grid.shape)

        max_value = max(max_val)
        min_value = min(min_val)
        grid_margins2 = [np.linspace(min_value-2, max_value+2, nr_points)] * len(df.columns)
        grid2 = np.stack(np.meshgrid(*grid_margins2), -1).reshape(-1, len(grid_margins2))
        
        # print(grid2.shape)
        
        return grid2
    
    def crossexam_dr(self, KD, d1, d2, g1, g2, embedding_type = None):
        """
        PARAMETERS
        ----------
        
        KD : density estimation for each individual: generated and control.
        
        name : the name of the cross examination, used for progressbar.
        
        d1 : str
            data source 
        
        d2 : str
            data source
        
        g1 : str
            group
        
        g2 : str
            group
            
        embedding_type : str
            
        
        RETURNS
        -------
        cross : DataFrame
            KL divergence scores between each individuals
        """

        
        left = np.sort(list(KD[d1][g1].keys()))
        right = np.sort(list(KD[d2][g2].keys()))
        

        cross = pd.DataFrame(columns=left.astype("str"), index=right.astype("str"), dtype="int64")
        
        name = ""
        
        if embedding_type:
            name += embedding_type + " "
        
        name += "{}-{} vs {}-{}".format(d1, g1, d2, g2)
        
        p = cu.ProgressBar(endVal=len(right)*len(left), name = name)
        while(p.move()):
            for i in right:
                for j in left:
                    cross.loc[str(i),str(j)] = self.KL_between_two_patients(KD[d1][g1][j],\
                                                                      KD[d2][g2][i])
        return cross 
    
    def kl_between_pairs(self, original, generated, embedding_type="", original_density=None):
        """
        PARAMETERS
        ----------
        original : a DataFrame
            containing both the control and diseased groups of the original data, can be embedded
 
        generated: a DataFrame 
            containing both the control and diseased groups of the generated data, can be embedded

        embeddingType: String 
            Used for progress bar.
        
        original_density : a dictionary
            contains the density estimation of the original groups

        RETURNS
        -------
        pairwise_kl: a list of dataframes. each dataframe contains a kl between two groups.

        """

        c = "control"
        d = "diseased"
        o = "original"
        g = "generated"
        
        nr_points = 120 if self.dim_red else 10
        
        # compute grid
        grid = self.computeGrid_dr(df = original[original.columns.difference(["id","group"])], nr_points=nr_points)
        print("grid computation done")
            
        # estimate density of both group (c and d) on the grid
        org = original.groupby("group")
        gen = generated.groupby("group")

        KD = { o: { c: original_density[c] if original_density else self.computeDensity_dr(org.get_group(c),grid), 
                    d: original_density[d] if original_density else self.computeDensity_dr(org.get_group(d),grid)
                  },
               g: { c: self.computeDensity_dr(gen.get_group(c),grid),
                    d: self.computeDensity_dr(gen.get_group(d),grid)
                  }
             }
        print("density estimation done")


        # compute the 4 kl-loss between matrices
        result = { 
            "originalControl vs generatedControl": self.crossexam_dr(KD, o, g, c, c, embedding_type),
            "originalDiseased vs generatedDiseased": self.crossexam_dr(KD, o, g, d, d, embedding_type),
            "originalControl vs originalDiseased":  self.crossexam_dr(KD, o, o, c, d, embedding_type),
            "generatedControl vs generatedDiseased":  self.crossexam_dr(KD, g, g, c, d, embedding_type)
        }

        # return the original_density to avoid computing it next time.
        orginal_density = {c: KD[o][c], d: KD[o][d]}
        
        return result, original_density
```

```python
def embedded_patients(df, embeddor, pre_embedded=None):
    # embedd generated patients
    e = embeddor(df[df.columns.difference(["id", "group"])], pre_embedded)
    # create dataframes from the embedded generated patients
    p = pd.DataFrame(e, index=df.index, columns=["fstAxis","sndAxis"])
    p["id"] = df.id
    return p
```

```python
def join(embedding):
    ab = np.concatenate((embedding["originalControl vs generatedControl"].values,\
                         embedding["originalDiseased vs generatedDiseased"].values),axis=1)
    cd = np.concatenate((embedding["originalControl vs originalDiseased"].values,\
                         embedding["generatedControl vs generatedDiseased"].values),axis=1)
    return np.concatenate((ab,cd),axis=0)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from matplotlib.colors import ListedColormap
```

```python
def plot_heatmap(embedding, title):
    f, axes = plt.subplots(2,2, sharey=False, figsize = (20,12))
    [[ax1,ax2],[ax3,ax4]] = axes

    g1 = sns.heatmap(embedding["originalControl vs generatedControl"], cmap = "coolwarm_r", cbar=False, ax=ax1)
    g1.set_ylabel('Generated Control')
    g1.set_xlabel('Original Control')
    g1.axes.xaxis.set_visible(False)

    g2 = sns.heatmap(embedding["generatedControl vs generatedDiseased" ], cmap="coolwarm_r", cbar=False, ax=ax2)
    g2.set_ylabel( 'Generated Control')
    g2.set_xlabel( 'Generated Diseased')
    g2.axes.xaxis.set_visible(False)
    g2.axes.yaxis.set_visible(False)

    g3 = sns.heatmap(embedding["originalControl vs originalDiseased"], cmap="coolwarm_r", cbar=False, ax=ax3)
    g3.set_ylabel('Original Diseased')
    g3.set_xlabel('Original Control')

    g4 = sns.heatmap(embedding["originalDiseased vs generatedDiseased"].T, cmap="coolwarm_r", cbar=False, ax=ax4)
    g4.set_ylabel('Original Diseased')
    g4.set_xlabel('Generated Diseased')
    g4.axes.yaxis.set_visible(False)

    f.suptitle(title, fontsize = 'x-large', fontweight = 'bold', ha="left", va="center", y=0.91, x=0.33)
    f.subplots_adjust(wspace=0.05, hspace=0.05)
    mappable = g4.get_children()[0]
    f.colorbar(mappable, ax = axes)
    
```

## DD-loss

compute dd-loss between a and b (their "cross loss" illustrated with the heatmap)
1. compute average kl between a[control] and b[control] (sim_c)
2. compute average kl between a[diseased] and b[diseased] (sim_d)
3. compute average kl between a[control] and b[diseased] (i)
4. compute average kl between a[control] and b[diseased] (ii)
5. compute absolute difference between (i) (ii) (div)
6. return sim_c + sim_d + div

```python
def ddloss(kl):
    """
    PARAMTERS
    ---------
    embedding : dict
        The kl-divergence scores necessary to compute the double date loss score.
    
    RETURNS
    -------
    loss : float
        The double date loss score.
    """
    OC_GC = kl["originalControl vs generatedControl"]
    OD_GD = kl["originalDiseased vs generatedDiseased"]
    GC_GD = kl["generatedControl vs generatedDiseased"]
    OC_OD = kl["originalControl vs originalDiseased"]
    
    div =  abs(GC_GD.mean().mean() - OC_OD.mean().mean())
    
    sim_c = abs(OC_GC.mean().mean())
    
    sim_d = abs(OD_GD.mean().mean())
    
    return sim_c + sim_d + div
```

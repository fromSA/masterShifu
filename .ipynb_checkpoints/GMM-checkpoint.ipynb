```python
class GMM_MODEL():
     
    def __init__(self):
        pass
    
    def convert_group(self, df):
        cond = pd.get_dummies(df["group"]).astype("float32")

        data = df[df.columns.difference(["id","group"])]
        data = np.concatenate((data, cond), axis=1)

        return cond, data
    
    def fit(self, data, latent_dim, epochs, batch_size, optimizer):
        pass


    def save(self, filepath):
        # save the model
        with open(filepath, 'wb') as f:
            pickle.dump(self.model, f)

        
    def load_generator(file):
        """
        PARAMETERS
        ----------
        file: a pickled sklearn gaussian mixture model object

        RETURNS
        -------
        unpickel: an unpickeled object

        """

        # and later you can load it
        
    
    def load(self, filepath):
        # load the model
        # set self.model
        
        # TODO if filepath exists.
        with open(filepath, 'rb') as f:
            self.model = pickle.load(f)
        
        
    def generate_patients(self, nr_markers=12, nr_cells = 20000, nr_patients = 20, column_names=None, group=None):
        """
        PARAMETERS:
        ----------
        nr_markers : int
            nr of markers

        nr_cells : int
            number of cells per patient
        
        nr_patients : int
            number of patients to generate
        
        column_names : 
            Dataframe.columns, names for the markers
        
        group : str
            "control" or "diseased"

        RETURNS:
        -------
        patients : dataframe
            list of patients, each patient with `sample_size` cells.
        """
        if self.model:
            
            patients = np.empty(shape=(nr_patients* nr_cells, nr_markers))
            p_id = np.empty(shape=nr_patients*nr_cells, dtype="int32")
            
            for i in range(nr_patients):
                p_id[nr_cells*i : nr_cells*(i+1)] = np.full(shape=(nr_cells), fill_value=i+1, dtype="int32")
                patients[nr_cells*i : nr_cells*(i+1)] = self.model.sample(nr_cells)[0]
            
            patients_df = pd.DataFrame(patients, columns=column_names)
            patients_df["id"] = p_id
            patients_df["group"] = group
            
            return patients_df    
        
        
```
